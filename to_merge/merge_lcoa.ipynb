```python
import pandas as pd
import os
import glob

# Define the directory containing the CSV files
directory = os.path.join(os.getcwd()#, 'to_merge'
                                                )
print(directory)
# Get a list of all CSV files in the directory
files = glob.glob(os.path.join(directory, '2050_lcoa_global_20231105*.csv'))

# Initialize a dictionary to store DataFrames
dfs = {}
files
```

```python



# Iterate over the CSV files, read each file, and store it in the dictionary
for file in files:
    df = pd.read_csv(file)
    filename = os.path.basename(file)
    name = filename.split('_')[4]  # Extract the name part from the file name
    dfs[name] = df

dfs.keys()
```

```python

dfs['120to140mp.csv']
```

```python

# Concatenate all DataFrames in the dictionary
data = pd.concat(dfs.values(), ignore_index=True)

# Sort the DataFrame by Increasing Longitude and decreasing Latitude
data.sort_values(by=['Longitude', 'Latitude'], ascending=[True, False], inplace=True)

# Save the concatenated and sorted data to a single CSV file
data.to_csv(os.path.join(directory, '2050_lcoa_global.csv'), index=False)
```
